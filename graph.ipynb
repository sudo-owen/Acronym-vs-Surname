```python
import json
import matplotlib.pyplot as plt

with open("author.json", "rb") as file:
  author = json.load(file)

with open("title.json", "rb") as file:
  title = json.load(file)

author_counts = {}
title_counts = {}

def make_counts(data, d):
  for k in data:
    count = data[k]
    try:
      d[count] = d[count] + 1
    except KeyError:
      d[count] = 1

make_counts(author, author_counts)
make_counts(title, title_counts)
```

```python
def make_pie(labels, sizes, title):
    plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
    fig, ax = plt.subplots()
    wedges, texts, autotexts = ax.pie(sizes, 
            labels=labels, 
            colors=['forestgreen','firebrick'],
            autopct='%1.1f%%',
            textprops=dict(color="w"))
    ax.legend(wedges, labels,
             title="Collisions",
             loc="center left",
             bbox_to_anchor=(1, 0, 0.5, 1))
    plt.setp(autotexts, size=8, weight="bold")
    ax.axis('equal')
    ax.set_title(title)
    plt.show()

labels = ['No collision', 'â‰¥ 1 collision']
sizes = [40260/41000, 740/41000]
title = "Collisions using title acronyms"
make_pie(labels, sizes, title)
```

```python
def make_bar(labels, sizes, title):
    fig, ax = plt.subplots()
    ax.barh(labels, sizes, height=1)
    ax.set_title(title)
    ax.invert_yaxis()
    ax.set_xlabel("Counts")
    ax.set_ylabel("Number of collisions")
    ax.tick_params(axis='y', which='major', pad=10)
    for i, v in enumerate(sizes):
        ax.text(v + 3, i + .25, str(v), color='black', fontweight='bold')
    plt.show()
```

```python
plt.rcParams["figure.figsize"] = (11,22)

make_bar([str(i) for i in list(author_counts.keys())], 
         list(author_counts.values()),
             'Distribution of collision count using last names')
```

```python
plt.rcParams["figure.figsize"] = (8, 3)

make_bar([str(i) for i in list(title_counts.keys())], 
         list(title_counts.values()),
             'Distribution of collision count using title acronyms')
```
